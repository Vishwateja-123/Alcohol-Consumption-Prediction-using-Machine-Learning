```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Sml_data.csv')
df
```

```python
df.isnull()
```

# Changing String to Integer Datatype

```python
df['DRK_YN']=df['DRK_YN'].replace({'Y':1,'N':0})
df['sex'] = df['sex'].replace({'Male':1,'Female':0})
df
```

```python
df.dtypes
```

```python
target=df['DRK_YN']
target
```

```python
features=df.drop(['DRK_YN'],axis=1)
features
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

# Perceptron Before PCA

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
model=Perceptron()
model.fit(X_train ,y_train)
y_pred=model.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```

```python
from sklearn.metrics import confusion_matrix
c=confusion_matrix(y_test,y_pred)
c
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

# Logistic Regression Before PCA

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model=LogisticRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```

```python
from sklearn.metrics import confusion_matrix
c=confusion_matrix(y_test,y_pred)
c
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

## SVM Before PCA

```python
from sklearn.svm import SVC
model=SVC()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```

```python
from sklearn.metrics import confusion_matrix
c=confusion_matrix(y_test,y_pred)
c
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

## KNN Before PCA

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)
y_pred = knn_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("KNN Accuracy:", accuracy)
```

# PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=6)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

# KNN After PCA

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
knn_model_pca = KNeighborsClassifier(n_neighbors=5)
knn_model_pca.fit(X_train_pca, y_train)
y_pred_pca = knn_model_pca.predict(X_test_pca)
accuracy_pca = accuracy_score(y_test, y_pred_pca)
print("KNN Accuracy after PCA:", accuracy_pca)
```

# SVM After PCA

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
model = SVC()
model.fit(X_train_pca, y_train)
y_pred = model.predict(X_test_pca)
accuracy = accuracy_score(y_test, y_pred)
print("SVC Accuracy with PCA:", accuracy)
print(classification_report(y_test, y_pred))
```

# Logistic Regression After PCA

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()
model.fit(X_train_pca, y_train)
y_pred = model.predict(X_test_pca)
accuracy = accuracy_score(y_test, y_pred)
print("Logistic Regression Accuracy with PCA:", accuracy)
print(classification_report(y_test, y_pred))
```

# Perceptron After PCA

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
model = Perceptron()
model.fit(X_train_pca, y_train)
y_pred = model.predict(X_test_pca)
accuracy = accuracy_score(y_test, y_pred)
print("Perceptron Accuracy with PCA:", accuracy)
print(classification_report(y_test, y_pred))
```

# Resampling

```python
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
for i in range(0, 10):
  xr, yr = resample(features, target)
  x_train, x_test, y_train, y_test = train_test_split(xr, yr, test_size=0.2)
```

## Logistic Regression After PCA

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()
accuracies_logistic = []
for i in range(0, 10):
  xr, yr = resample(features, target)
  x_train, x_test, y_train, y_test = train_test_split(xr, yr, test_size=0.2)
  model = LogisticRegression()
  model.fit(x_train, y_train)
  y_pred = model.predict(x_test)
  accuracy = accuracy_score(y_test, y_pred)
  accuracies_logistic.append(accuracy)

  print(f"LogisticRegression (Iteration {i+1}): {accuracy}")
```

# Perceptron After PCA

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
accuracies_perceptron = []
for i in range(0, 10):
  xr, yr = resample(features, target)
  x_train, x_test, y_train, y_test = train_test_split(xr, yr, test_size=0.2)
  model = Perceptron()
  model.fit(x_train, y_train)
  y_pred = model.predict(x_test)
  accuracy = accuracy_score(y_test, y_pred)
  accuracies_perceptron.append(accuracy)
  print(f"Perceptron Accuracy (Iteration {i+1}): {accuracy}")
```

# KNN After PCA

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
accuracies_knn = []
for i in range(0, 10):
  xr, yr = resample(features, target)
  x_train, x_test, y_train, y_test = train_test_split(xr, yr, test_size=0.2)
  model = KNeighborsClassifier()
  model.fit(x_train, y_train)
  y_pred = model.predict(x_test)
  accuracy = accuracy_score(y_test, y_pred)
  accuracies_knn.append(accuracy)
  print(f"KNN Accuracy (Iteration {i+1}): {accuracy}")
```

# SVM After PCA

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
accuracies_svc = []
for i in range(0, 10):
  xr, yr = resample(features, target)
  x_train, x_test, y_train, y_test = train_test_split(xr, yr, test_size=0.2)
  model = SVC()
  model.fit(x_train, y_train)
  y_pred = model.predict(x_test)
  accuracy = accuracy_score(y_test, y_pred)
  accuracies_svc.append(accuracy)
  print(f"SVC Accuracy (Iteration {i+1}): {accuracy}")
```

# Plotting Graphs Of All Models

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.arange(1,11)
print(a)
print(accuracies_logistic)
print(accuracies_perceptron)
print(accuracies_knn)
print(accuracies_svc)
plt.plot(range(1, 11), accuracies_logistic, label='Logistic Regression')
plt.plot(range(1, 11), accuracies_perceptron, label='Perceptron')
plt.plot(range(1, 11), accuracies_knn, label='KNN')
plt.plot(range(1, 11), accuracies_svc, label='SVC')
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.title('Model Accuracies Over Iterations')
plt.legend()
plt.show()
```
